### Data manupulation / Data cleaning

#### - Extracting year from df3 Time period column as there is complete date given which is not required only year is required so we extracted year

```python
df3 = pd.read_csv('Netflix_Customer _Subscription.csv' , lineterminator = '\n')

df3['Time Period'] = pd.to_datetime(df3['Time Period'])
print(df3['Time Period'].dtype)

df3['Time Period'] = df3['Time Period'].dt.year
df3.head()
```

#### - Replacing Nan Values

```python
df1.fillna('Unknown/NA' , inplace=True)
df1.head()
```

#### - Dropping unwanted columns

```python
cols = ['cast' , 'director' , 'date_added' , 'description']
df1.drop(cols , axis =1 , inplace = True)
df1.head()
```

#### - Renaming columns

* Now we will split genre into a list and then explode our data frame so then we will have one genre per column

* Doing this will increse our number od columns and rows in df1 data frame


```python
df1['listed_in'] = df1['listed_in'].str.split(', ')
df1 = df1.explode('listed_in').reset_index(drop = True)
df1.head()

```

### Que1 ) Netflix’s global revenue and subscriber count changes over the years.

```python
# Netflix global revenue count over years -

df4 = pd.read_excel("Netflix Revenue and Usage Statistics.xlsx")

#plt.figure(figsize=(10,6))


plt.bar(x = df4['Year'] , height = df4['Revenue ($bn)'] , color = 'darkred' , label = 'Netflix revenue over years' , edgecolor = 'black')
plt.xlabel('Years' , color = 'darkred')
plt.ylabel('Revenue ($bn)' , color = 'darkred')
plt.legend()
plt.title('Netflix revenue over years 2012-2022' ,  color = 'darkred')
plt.tight_layout()
plt.grid(linestyle = ':' , linewidth = '1')

plt.show()



# Netflix global Subscriber count over years -
plt.barh(
    y=df3['Time Period'], 
    width=df3['Subscribers\r'], 
    color='darkred', 
    edgecolor='black'
)
plt.xlabel("Subscribers (in millions)" , color = 'darkred')
plt.ylabel("Time Period" , color = 'darkred')
plt.title("Subscribers Count Over Years" , color = 'darkred')
plt.grid(linestyle = ':' , linewidth = '1')
plt.tight_layout()
plt.show()



```

#### Insights and Key trends

##### Revenue growth over years 2012-2022

###### Thunderous Growth (2012–2018)
* Netflix’s revenue surged from 3.6B to 15.8B, representing roughly fourfold growth within six years.

###### Steady Climb Continues (2019–2021)
* Revenue climbed from 20.16B in 2019 to 29.70B in 2021.

##### Subscribers growth over years 2012-2022
* 2012–2019: Strong, consistent growth with subscriber base surging from ~26 million to ~151 million.
* 2020: A clear spike, ending the year near 193 million subscribers.
* 2021–2022: Growth slowed — by 2022 Netflix reached about ~231 million subscribers, indicating a more mature stage in its lifecycle.
